```python
# Listing the average score per school type
%sql SELECT "Elementary, Middle, or High School", AVG(SAFETY_SCORE) AS AVERAGE_SAFETY_SCORE FROM CHICAGO_PUBLIC_SCHOOLS\
GROUP BY "Elementary, Middle, or High School"
```

```python
# Listing the top 5 community areas with the % of scores below the poverty line
%sql SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY FROM CHICAGO_CENSUS_DATA\
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC\
LIMIT 5
```

```python
# Looking at the community areas with the largest number of crimes. (None represents the cells that do not have a community number. Hence, we can say that these data are distorted, because in case all the empty cells correspond to community area # 23, then community area 23 would be the community area with the largest number of crimes. Since the lowest number of crimes is 1 and the largets number is 43, hence if all the "Nones" correspond to one community area then that community area would be ranked number 1. And many other options can emerge since most likely the "Nones" do not correpsond to one single community) )
%sql SELECT COMMUNITY_AREA_NUMBER, COUNT (*) AS NUMBER_OF_CRIMES FROM CHICAGO_CRIME_DATA\
GROUP BY COMMUNITY_AREA_NUMBER ORDER BY NUMBER_OF_CRIMES DESC 
```

```python
#Finding the community area with the largest number
%sql SELECT COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER FROM CHICAGO_CENSUS_DATA WHERE \
HARDSHIP_INDEX =(SELECT MAX(HARDSHIP_INDEX) AS MAX_HARDSHIP_INDEX FROM CHICAGO_CENSUS_DATA)
```

```python
# Selecting the community area name for which the crime rate is the highest using a sub query
%sql SELECT COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER \
FROM CHICAGO_CENSUS_DATA WHERE COMMUNITY_AREA_NUMBER \
IN(SELECT COMMUNITY_AREA_NUMBER FROM CHICAGO_CRIME_DATA GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT (*) DESC LIMIT 1)
```
